```python
data= data.fillna(data.median(numeric_only=True)) # fill the missing values with median for numeric columns
data.head()
```

```python
numeric_columns = ['goal', 'pledged', 'usd pledged', 'usd_goal_real', 'backers']
for col in numeric_columns:  ## this is for remove the outliers from the numeric columns
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    LOWER_bound = Q1 - 1.5 * IQR
    UPPER_bound = Q3 + 1.5 * IQR
    data = data[(data[col] >= LOWER_bound) & (data[col] <= UPPER_bound)]
```

```python
categorical_columns=['category', 'main_category', 'currency', 'country'] ## here we select the categorical columns
one_hot_encoding=pd.get_dummies(data , columns=categorical_columns , drop_first=True) ## converts categorical columns to one-hot-encoded format (true , false)
binary_columns= one_hot_encoding.select_dtypes(include=['bool']).columns ## identifies columns with boolean data types
one_hot_encoding[binary_columns]=one_hot_encoding[binary_columns].astype(int) ## convert the (true,false) to binary --> [0,1]
one_hot_encoding
```

```python
numeric_data = one_hot_encoding[numeric_columns]
min_max = (numeric_data - numeric_data.min()) // (numeric_data.max() - numeric_data.min())
min_max.head()
```

```python
#one_hot_encoding.drop(columns=numeric_columns): This removes the original numeric columns from a DataFrame that has categorical data converted to numbers (one-hot encoded).
#min_max: this has the numeric columns but scaled to be between 0 and 1 (normalized).
final_data = pd.concat([one_hot_encoding.drop(columns=numeric_columns), min_max], axis=1)
final_data.head()
```

```python
features_col=['category', 'main_category', 'currency', 'deadline', 'goal',
       'pledged', 'state', 'backers', 'country', 'usd pledged',
       'usd_pledged_real', 'usd_goal_real']
X=data[features_col]
y=data.state ## target variable
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size=0.25 , random_state=16) 
# test_size=0.25 means that 25% of the data will be used for testing and the remaining 75% will be used for training
```

```python
logistic_regration = LogisticRegression(random_state=16 ,  max_iter=5000 , solver='liblinear')
```

```python
logistic_regration.fit(X_train, y_train) 
# fit() Method: This method is used to train the logistic regression model using the training data
```

```python
predictions = logistic_regration.predict(X_test)
```

```python
confusion_matrix = metrics.confusion_matrix(y_test , predictions)
confusion_matrix
```

```python
class_names=[0,1] # name  of classes
x, y = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# now we will create a heatmap plot
sns.heatmap(pd.DataFrame(confusion_matrix), annot=True, cmap="YlGnBu" ,fmt='g') 
# annot --> to see the numbers in the boxs of the heatmap
# cmap --> this is the colors : (yl : yellow ) (gn : green) (bu : blue) Colors change from yellow to blue with higher values, highlighting data distribution
# fmt --> displays numbers directly, large or small without editing anything
y.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
(0.5,257.44,'Predicted label'); # the semicolon used to delete the text which will be printed up the plot
```

```python
target_names = ['class 0', 'class 1', 'class 2', 'class 3', 'class 4']
print(classification_report(y_test, predictions, target_names=target_names))
```
